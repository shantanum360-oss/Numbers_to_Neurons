```
import numpy as np

# -----------------------------
# XOR Dataset
# -----------------------------
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

y = np.array([
    [0],
    [1],
    [1],
    [0]
])

# -----------------------------
# ReLU Activation
# -----------------------------
def relu(z):
    return np.maximum(0, z)

def relu_derivative(z):
    return (z > 0).astype(float)

# -----------------------------
# Sigmoid Activation
# -----------------------------
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# -----------------------------
# Network Architecture
# Input -> Hidden -> Output
# 2 -> 4 -> 1
# -----------------------------
np.random.seed(42)

W1 = np.random.randn(2, 4) * 0.1
b1 = np.zeros((1, 4))

W2 = np.random.randn(4, 1) * 0.1
b2 = np.zeros((1, 1))

learning_rate = 0.1
epochs = 10000

# -----------------------------
# Training Loop
# -----------------------------
for epoch in range(epochs):

    # ===== Forward Pass =====

    Z1 = X @ W1 + b1
    A1 = relu(Z1)

    Z2 = A1 @ W2 + b2
    A2 = sigmoid(Z2)

    # ===== Loss Computation =====
    loss = np.mean((A2 - y) ** 2)

    # ===== Backward Pass =====

    m = X.shape[0]

    # Output layer gradients
    dA2 = 2 * (A2 - y) / m
    dZ2 = dA2 * A2 * (1 - A2)

    dW2 = A1.T @ dZ2
    db2 = np.sum(dZ2, axis=0, keepdims=True)

    # Hidden layer gradients
    dA1 = dZ2 @ W2.T
    dZ1 = dA1 * relu_derivative(Z1)

    dW1 = X.T @ dZ1
    db1 = np.sum(dZ1, axis=0, keepdims=True)

    # ===== Gradient Descent Update =====

    W2 -= learning_rate * dW2
    b2 -= learning_rate * db2

    W1 -= learning_rate * dW1
    b1 -= learning_rate * db1

    if epoch % 1000 == 0:
        print(f"Epoch {epoch}, Loss = {loss:.6f}")

# -----------------------------
# Final Predictions
# -----------------------------
Z1 = X @ W1 + b1
A1 = relu(Z1)

Z2 = A1 @ W2 + b2
predictions = sigmoid(Z2)

print("\nPredictions:")
print(predictions)



#**Explanation**

"""For my modification, I replaced the hidden layer activation function from the sigmoid function to the ReLU (Rectified Linear Unit) activation function.

Mathematically, the sigmoid function maps every input to a value between 0 and 1 using the formula σ(x) = 1/(1 + e⁻ˣ). In contrast, the ReLU function outputs 0 for negative inputs and returns the input itself for positive inputs, i.e., ReLU(x) = max(0, x). During backpropagation, the derivative of ReLU is 1 for positive inputs and 0 for non-positive inputs, which makes the gradient calculations simpler.

I expected ReLU to behave differently because sigmoid activations can become saturated when their inputs are very large or very small, resulting in extremely small gradients. This can slow down learning. ReLU generally allows gradients to propagate more effectively for positive inputs, often leading to faster training and better convergence compared to a network that uses sigmoid activations throughout."""
```
