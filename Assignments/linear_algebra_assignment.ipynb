### Q2 Part (iii) — NumPy Function

Write a NumPy function that:

````python
import numpy as np

# Input — do not modify
A = np.array([[2, 0], [0, 3]])
v1 = np.array([1, 0])
v2 = np.array([0, 1])
v3 = np.array([1, 1])
v4 = np.array([-2, 4])

# Function
def transform_vector(matrix, vector):
    transformed = matrix @ vector

    # Describe scaling
    x_scale = matrix[0, 0]
    y_scale = matrix[1, 1]

    print(f"x-component scaled by {x_scale}, y-component scaled by {y_scale}")
    print(f"Original vector: {vector}")
    print(f"Transformed vector: {transformed}\n")

    return transformed

# Testing on all vectors
transform_vector(A, v1)
transform_vector(A, v2)
transform_vector(A, v3)
transform_vector(A, v4)
````

## Q3 — Tiny Recommendation System

Each user's taste is represented as a vector of 4 scores across 4 genres — **action, romance, thriller, comedy** — all between 0 and 1.

```
User A: [0.9, 0.1, 0.8, 0.2]
User B: [0.1, 0.9, 0.2, 0.8]
User C: [0.8, 0.2, 0.9, 0.1]
User D: [0.5, 0.5, 0.5, 0.5]
```

### Q3 Part (i) — Similarity Matrix

Compute the **full similarity matrix** — every user's similarity score against every other user — using only a **single NumPy operation** (no loops).

What linear algebra concept makes this possible in one line?

````python
# Input — do not modify
users = np.array([
    [0.9, 0.1, 0.8, 0.2],  # User A
    [0.1, 0.9, 0.2, 0.8],  # User B
    [0.8, 0.2, 0.9, 0.1],  # User C
    [0.5, 0.5, 0.5, 0.5],  # User D
])
````

````python
import numpy as np

# Input — do not modify
users = np.array([
    [0.9, 0.1, 0.8, 0.2],  # User A
    [0.1, 0.9, 0.2, 0.8],  # User B
    [0.8, 0.2, 0.9, 0.1],  # User C
    [0.5, 0.5, 0.5, 0.5],  # User D
])


similarity_matrix = users @ users.T

print(similarity_matrix)
````

**What linear algebra concept makes this possible in one line?**


The linear algebra concept is the dot product implemented through matrix multiplication.NumPy multiplies each user vector with every other user vector simultaneously.

### Q3 Part (iii) — The Problem with Raw Dot Products

A senior engineer tells you:
> *"Raw dot products are a bad measure of similarity — you need to normalize."\*

1. What is the **problem** with raw dot products?
2. Construct a **specific example using NumPy** that demonstrates the flaw.
3. Fix it.

> **Hint:** Take a user with scores `[1.8, 0.2, 1.6, 0.4]`

Problem with raw dot products

Raw dot products depend on both:

the direction of the vectors (taste similarity)
the magnitude of the vectors (overall score size)

This means a user who simply gives larger ratings can appear “more similar” even if their preferences are exactly the same.
So raw dot products confuse preference pattern with rating scale.

````python
import numpy as np

A = np.array([0.9, 0.1, 0.8, 0.2])
scaled_A = np.array([1.8, 0.2, 1.6, 0.4])


dot_product = A @ scaled_A

print("Raw dot product:", dot_product)
````

### Q3 Part (iv) — Most Similar Users

After normalizing, which **two users are most similar**?

Does that match your intuition from just reading the vectors?

What does it mean in plain language — what do these two users have in common?

> **Note:** A similarity score is just a number that tells you **how alike two things are**.

After normalization, the most similar users are **User A and User C**.

This matches intuition from reading the vectors:

* User A: `[0.9, 0.1, 0.8, 0.2]`
* User C: `[0.8, 0.2, 0.9, 0.1]`

Both users strongly prefer:

* **action**
* **thriller**

and both dislike:

* **romance**
* **comedy**

Their scores are almost mirror images of each other, so their vectors point in nearly the same direction after normalization.

### Plain language meaning

These two users enjoy the same kinds of movies.
They both like intense, exciting genres and are less interested in romantic or comedic content.

So, in a recommendation system, if User A enjoys a movie, there is a high chance User C would also enjoy it.


````python
import numpy as np


users = np.array([
    [0.9, 0.1, 0.8, 0.2],  # User A
    [0.1, 0.9, 0.2, 0.8],  # User B
    [0.8, 0.2, 0.9, 0.1],  # User C
    [0.5, 0.5, 0.5, 0.5],  # User D
])


norms = np.linalg.norm(users, axis=1, keepdims=True)
normalized_users = users / norms


similarity_matrix = normalized_users @ normalized_users.T

print("Similarity Matrix:\n")
print(similarity_matrix)


np.fill_diagonal(similarity_matrix, -1)


max_index = np.unravel_index(
    np.argmax(similarity_matrix),
    similarity_matrix.shape
)

user_names = ["A", "B", "C", "D"]

print("\nMost similar users:")
print(f"User {user_names[max_index[0]]} and User {user_names[max_index[1]]}")

print("\nSimilarity score:")
print(similarity_matrix[max_index])
````
